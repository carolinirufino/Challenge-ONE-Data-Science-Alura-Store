### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Função para calcular o faturamento de cada loja
def calcular_faturamento(loja, nome):
    total = loja["Preço"].sum()
    print(f"Faturamento {nome}: R$ {total:,.2f}")
    return total

faturamento_loja1 = calcular_faturamento(loja, "Loja 1")
faturamento_loja2 = calcular_faturamento(loja2, "Loja 2")
faturamento_loja3 = calcular_faturamento(loja3, "Loja 3")
faturamento_loja4 = calcular_faturamento(loja4, "Loja 4")
```

```python
# Calculando o faturamento total das lojas
faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4
print(f"Faturamento Total: R$ {faturamento_total:,.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x=["Loja 1", "Loja 2", "Loja 3", "Loja 4"], height=[faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4])
plt.xlabel("Lojas")
plt.ylabel("Faturamento")
plt.title("Faturamento por Loja")
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
vendas = [
    {"loja": "Loja 1", "categoria": "Roupas", "valor": 120, "avaliacao": 4},
    {"loja": "Loja 1", "categoria": "Alimentos", "valor": 80, "avaliacao": 5},
    {"loja": "Loja 2", "categoria": "Eletrônicos", "valor": 500, "avaliacao": 3},
    {"loja": "Loja 2", "categoria": "Roupas", "valor": 200, "avaliacao": 2},
    {"loja": "Loja 3", "categoria": "Alimentos", "valor": 150, "avaliacao": 4},
    {"loja": "Loja 3", "categoria": "Roupas", "valor": 100, "avaliacao": 5},
    {"loja": "Loja 4", "categoria": "Eletrônicos", "valor": 700, "avaliacao": 4},
    {"loja": "Loja 4", "categoria": "Alimentos", "valor": 120, "avaliacao": 3},
]
```

```python
# Dicionários para somar avaliações e contar quantas tem
soma_avaliacoes = {}
qtd_avaliacoes = {}

# Loop pra preencher
for venda in vendas:
    loja = venda["loja"]
    avaliacao = venda["avaliacao"]

    if loja not in soma_avaliacoes:
        soma_avaliacoes[loja] = 0
        qtd_avaliacoes[loja] = 0

    soma_avaliacoes[loja] += avaliacao
    qtd_avaliacoes[loja] += 1

# Calculando a média
media_avaliacoes = {}

for loja in soma_avaliacoes:
    media = soma_avaliacoes[loja] / qtd_avaliacoes[loja]
    media_avaliacoes[loja] = round(media, 2)  # arredondado com 2 casas

# Mostrando o resultado
for loja, media in media_avaliacoes.items():
    print(f"{loja}: Média de Avaliação = {media} estrelas")
```

```python
plt.figure(figsize=(8, 5))
plt.barh(list(media_avaliacoes.keys()), list(media_avaliacoes.values()), color="orchid")
plt.title("Média de Avaliação das Lojas")
plt.xlabel("Avaliação Média (1 a 5)")
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()

```

# 5. Frete Médio por Loja







```python
vendas = [
    {"loja": "Loja 1", "produto": "Camiseta", "valor": 120, "frete": 15, "avaliacao": 4},
    {"loja": "Loja 1", "produto": "Arroz", "valor": 80, "frete": 10, "avaliacao": 5},
    {"loja": "Loja 2", "produto": "Notebook", "valor": 500, "frete": 25, "avaliacao": 3},
    {"loja": "Loja 2", "produto": "Camiseta", "valor": 200, "frete": 20, "avaliacao": 2},
    {"loja": "Loja 3", "produto": "Arroz", "valor": 150, "frete": 12, "avaliacao": 4},
    {"loja": "Loja 3", "produto": "Camiseta", "valor": 100, "frete": 10, "avaliacao": 5},
    {"loja": "Loja 4", "produto": "Notebook", "valor": 700, "frete": 30, "avaliacao": 4},
    {"loja": "Loja 4", "produto": "Arroz", "valor": 120, "frete": 15, "avaliacao": 3},
]
```

```python
frete_por_loja = {}
contagem_por_loja = {}

for venda in vendas:
    loja = venda["loja"]
    frete = venda["frete"]

    if loja not in frete_por_loja:
        frete_por_loja[loja] = 0
        contagem_por_loja[loja] = 0

    frete_por_loja[loja] += frete
    contagem_por_loja[loja] += 1

# Calcular média
frete_medio_por_loja = {}

for loja in frete_por_loja:
    media = frete_por_loja[loja] / contagem_por_loja[loja]
    frete_medio_por_loja[loja] = round(media, 2)

print(frete_medio_por_loja)
```

```python
import matplotlib.pyplot as plt

lojas = list(frete_medio_por_loja.keys())
fretes = list(frete_medio_por_loja.values())

plt.figure(figsize=(8, 5))
plt.plot(lojas, fretes, marker='o', linestyle='-', color='darkorange')

for i, valor in enumerate(fretes):
    plt.text(i, valor + 0.5, f"R$ {valor}", ha='center', fontsize=9)

plt.title("🚚 Frete Médio por Loja")
plt.xlabel("Lojas")
plt.ylabel("Frete Médio (R$)")
plt.ylim(0, max(fretes) + 10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```
