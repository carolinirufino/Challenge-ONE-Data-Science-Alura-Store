### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Importando as bibliotecas necessárias.
import pandas as pd
import matplotlib.pyplot as plt

# Adicionando uma coluna 'Loja' com o nome da loja para cada DataFrame individual (loja1, loja2, loja3, loja4).
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenado os DataFrames das 4 lojas em um único DataFrame (df) para análise conjunta.
df = pd.concat([loja1, loja2, loja3, loja4])

# Calculando o faturamento total de cada loja, somando os valores da coluna 'Preço' filtrados pela loja específica.
faturamento_loja1 = df[df['Loja'] == 'Loja 1']['Preço'].sum()
faturamento_loja2 = df[df['Loja'] == 'Loja 2']['Preço'].sum()
faturamento_loja3 = df[df['Loja'] == 'Loja 3']['Preço'].sum()
faturamento_loja4 = df[df['Loja'] == 'Loja 4']['Preço'].sum()

# Mostrando o faturamento de cada loja.
print(f"Faturamento da Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento da Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento da Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento da Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Criação do gráfico
plt.figure(figsize=(8, 8))
plt.bar(lojas, faturamentos, color="pink")
plt.title("Faturamento Total por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Exibição do gráfico
plt.show()
```

# 2. Vendas por Categoria


```python

# Função para calcular e mostrar a contagem por categoria
def mostrar_contagem_por_categoria(df, nome_loja):
    categoria_contagem = df['Categoria do Produto'].value_counts()

    # Criando gráfico
    plt.figure(figsize=(10, 5))
    categoria_contagem.plot(kind='bar', color='lightpink')
    plt.title(f'Vendas por Categoria {nome_loja}')
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.show()

# Ajustando para cada loja
mostrar_contagem_por_categoria(loja1, "Loja 1")
mostrar_contagem_por_categoria(loja2, "Loja 2")
mostrar_contagem_por_categoria(loja3, "Loja 3")
mostrar_contagem_por_categoria(loja4, "Loja 4")
```

# 3. Média de Avaliação das Lojas

```python
# Calculando a média das avaliações por loja
media_avaliacao_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Criando um dicionário para armazenar e exibir o gráfico
media_avaliacao = {
    'Loja 1': media_avaliacao_loja1,
    'Loja 2': media_avaliacao_loja2,
    'Loja 3': media_avaliacao_loja3,
    'Loja 4': media_avaliacao_loja4
}

import matplotlib.pyplot as plt

# Criando o gráfico de linha
plt.figure(figsize=(8, 5))
plt.plot(list(media_avaliacao.keys()), list(media_avaliacao.values()), marker='o', linestyle='-', color='deeppink')
plt.title("Média de Avaliação das Lojas")
plt.xlabel("Lojas")
plt.ylabel("Média de Avaliação")
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

---



```python
def produtos_mais_menos_vendidos_juntos(df, nome_loja):
    # Conta os produtos
    contagem_produtos = df['Produto'].value_counts()

    # Top 10 mais vendidos e 10 menos vendidos (pode aumentar pra 10 se quiser)
    mais_vendidos = contagem_produtos.head(10)
    menos_vendidos = contagem_produtos.tail(10)

    # Cria novo DataFrame para plotagem
    comparacao_df = pd.DataFrame({
        'Produto': list(mais_vendidos.index) + list(menos_vendidos.index),
        'Quantidade': list(mais_vendidos.values) + list(menos_vendidos.values),
        'Tipo': ['Mais Vendido'] * len(mais_vendidos) + ['Menos Vendido'] * len(menos_vendidos)
    })

    # Plotando com seaborn
    import seaborn as sns
    plt.figure(figsize=(12,6))
    sns.barplot(data=comparacao_df, x='Produto', y='Quantidade', hue='Tipo', palette=['rosybrown', 'lightpink'])

    plt.title(f'Comparação: Produtos Mais e Menos Vendidos - {nome_loja}', fontsize=16)
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

# Chamadas para cada loja
produtos_mais_menos_vendidos_juntos(loja1, "Loja 1")
produtos_mais_menos_vendidos_juntos(loja2, "Loja 2")
produtos_mais_menos_vendidos_juntos(loja3, "Loja 3")
produtos_mais_menos_vendidos_juntos(loja4, "Loja 4")
```
